# Problem

Facial Recognition has been around for the past decades and has been applied to Criminal Databases, Traffic Cameras, Phone Security, etc. However, it does come with flaws despite the progress made in the field. Accuracy has been a major concern as it can misidentify a person or object and have negative consequences. Accuracy is affected by many factors, postures, illumination, colors, etc. We seek to improve accuracy when identifying images.

# Past and Current Solutions


---


* One of the methods employed in the past to enhance facial recognition accuracy has been the application of mathematical techniques. One of the many method involved comparing input face images with registered images using complex algorithms. This process utilized a database of images categorized into distinct classes, each with specific vector values representing facial features. For every image, a carefully defined distance function was implemented to measure the similarity between the feature vectors. By analyzing these distances, the system could accurately determine the class to which the input face belonged. This mathematical approach significantly contributed to the advancement of facial recognition technology, enabling more precise and reliable results to be obtained.

* One way of identification for Facial Recognition is "Support Vector Machine" which utilizes regression and classification techniques for analyzing images. By the use of vectors, computation time can be decreased and provide better accuracy by comparing numerical values.

* Facial features are seperated from the image, essentially it is broken into many squares/rectangles and then compared with each other throught multiple layers. Each square allows for comparison with multiple images, allowing for larger data size and better comparison techniques.

* The Quality of Images and Videos are factors to consider for classifying a person's face. The Model has difficulty identifying two images of the same person with different image quality. Illumination, Distance from camera, posture, etc. contribute to the misidentification. A general conclusion is to avoid low quality images to better train the model.

```python
#THIS CELL IS FOR LOADING IMAGES FROM THE LFW DATASET
#loads data from the lfw dataset on faces into a tuple of image tensors and labels
class LoadData():

  def __init__(self):

    #TRAINING DATASET
    self.Training = tuple()
    self.Validation = tuple()
    #LOAD FROM LFW DATASET
    self.numofImages = 256
    train_images = tfds.load('lfw', split='train', shuffle_files=True).take(self.numofImages) #number of images to import from dataset
    self.imageArray = [] #images
    labels = [] #labels

    for image in train_images:
      self.imageArray.append(tf.squeeze(tf.image.rgb_to_grayscale(image["image"])))
      labels.append(((image["label"]).numpy()).decode("utf-8"))
    train = tf.convert_to_tensor(self.imageArray)
    tensor = tf.constant(train)
    self.Training = (tensor, labels)
    print(labels)
    val_images = tfds.load('lfw', split='train', shuffle_files=True).take(self.numofImages) #number of images to import from dataset
    self.imageArray = [] #images
    labels = [] #labels

    for image in val_images:
      self.imageArray.append(tf.squeeze(tf.image.rgb_to_grayscale(image["image"])))
      labels.append(((image["label"]).numpy()).decode("utf-8"))
    train = tf.convert_to_tensor(self.imageArray)
    tensor = tf.constant(train)
    self.Validation = (tensor, labels)

#Loads the training/validation set
class FaceData(d2l.DataModule):
    #batch size is 64 because number of images imported from the dataset is 256
    def __init__(self, batch_size=64, resize=(64, 64)):
        self.save_hyperparameters()
        #Load the training set
        self.train = LoadData().Training
        self.val = LoadData().Validation
        self.numofClasses = LoadData().numofImages
        self.imageSize = (64, 64, 1)
        self.numofEpochs = 30
        self.numofBatches = 4
        #print(self.train[0])
FaceData()
```

# 4) Dataloader/Visualization
* Source: https://d2l.ai/chapter_linear-classification/image-classification-dataset.html

```python
#DATALOADER (modified function from textbook: https://d2l.ai/chapter_linear-classification/image-classification-dataset.html)
data = FaceData()
@d2l.add_to_class(FaceData)
def get_dataloader(self, train):
    data = self.train if train else self.val
    process = lambda X, y: (tf.expand_dims(X, axis=3) / 255,
                            tf.cast(y, dtype='string'))
    resize_fn = lambda X, y: (tf.image.resize_with_pad(X, *self.resize), y)
    shuffle_buf = len(data[0]) if train else 1
    #print(process(*data))
    return tf.data.Dataset.from_tensor_slices(process(*data)).batch(
        self.batch_size).map(resize_fn).shuffle(shuffle_buf)
X, y = next(iter(data.train_dataloader()))
```

```python
#TEXT LABELS (modified function from textbook: https://d2l.ai/chapter_linear-classification/image-classification-dataset.html)
@d2l.add_to_class(FaceData)
def text_labels(self, indices):
    labels = indices.numpy()
    return [labels[int(i)].decode("utf-8") for i in range(0, len(labels))]

#SHOW IMAGES (modified function from textbook: https://d2l.ai/chapter_linear-classification/image-classification-dataset.html)
def show_images(imgs, num_rows, num_cols, titles=None, scale=1.5):
    raise NotImplementedError

#VIZUALIZER (function from textbook: https://d2l.ai/chapter_linear-classification/image-classification-dataset.html)
@d2l.add_to_class(FaceData)
def visualize(self, batch, nrows=2, ncols=5, labels=[]):
    X, y = batch
    if not labels:
      labels = self.text_labels(y)
    d2l.show_images(tf.squeeze(X), nrows, ncols, titles=labels)
batch = next(iter(data.train_dataloader()))
data.visualize(batch)
```

# 5) Loading the Model

```python
#MODEL
@d2l.add_to_class(FaceData)
def get_module(self, image_shape, filter1, filter2, filter_shape, pool_shape, units, classes):
  model = Sequential()
  #adding a 2 dimensional convolutional layer
  model.add(Conv2D(filter1, filter_shape, activation='relu', input_shape=image_shape))
  #adding a 2 dimensional maxpooling layer
  model.add(MaxPooling2D(pool_shape))
  #adding another 2 dimensional convolutional layer
  model.add(Conv2D(filter2, filter_shape, activation='relu'))
  model.add(Flatten())
  #adding another dense layer for the classes with softmax activation
  model.add(Dense(classes, activation='softmax'))

  model.summary()
  return model
```

```python
#converts <string> labels to <int> indices
@d2l.add_to_class(FaceData)
def getlabels(self, y):
  labels = []
  for label in data.train[1]:
    if label not in labels:
      labels.append(label)
  #print(y)
  t_label = []
  for label in y:
    t_label.append(labels.index(label))
  t_label = tf.convert_to_tensor(t_label)
  return t_label
```

# 6) Running the Model

```python
#RUNNING THE MODEL
@d2l.add_to_class(FaceData)
def run(self):
  self.save_hyperparameters()
  #image shape: data.imageSize
  #filter1: 32
  #filter2: 32
  #filter shape: (3, 3)
  #pool shape: (2, 2)
  #units: 8
  #classes: (int) data.numofClasses
  self.model = data.get_module(data.imageSize, 32, 32, (3, 3), (2,2), 8, data.numofClasses)

  #setting the optimization algorithm to adam and the loss function to SCC and adding an accuracy metric
  self.model.compile(optimizer= 'adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])

  #load the training set (gets a batch from the dataloader with the batch size set in the  FaceData() class)
  #each time you call next(iter(...)), it loads a new batch
  #X is the images and y is the string labels for the images

  #runs multiple batches each with 25-30 epochs until it passes through the entire dataset
  for i in range(0, self.numofBatches):
    self.train_images, y = next(iter(data.train_dataloader()))
    self.train_labels = data.getlabels(y)
    self.test_images, z = next(iter(data.train_dataloader()))
    test_labels = data.getlabels(z)
    return self.model.fit(self.train_images, self.train_labels, batch_size=30, epochs=data.numofEpochs,validation_data=(self.test_images, test_labels))
model_data = (data.run())
```
